This first example of the basic usage of the Pyriod package will be demonstrated at the [TASC5/KASC12 workshop](https://web.mit.edu/tasc5/) in Boston 22-26 July 2019.  It shows how to reproduce the frequency analysis of a pulsating helium-atmosphere white dwarf presented in Bell et al. (A&A, submitted)

```python
%matplotlib notebook
import lightkurve as lk
from Pyriod import Pyriod
```

```python
lc = lk.search_lightcurvefile('257459955').download().get_lightcurve("PDCSAP_FLUX").flatten(2161).remove_outliers(5)
```

```python
pyriod = Pyriod(lc)
```

```python
pyriod.TimeSeries()
```

```python
pyriod.Periodogram()
```

```python
pyriod.Signals()
```

```python
pyriod.Log()
```
